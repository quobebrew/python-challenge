```python
# Python Script that analyzes records to calculate the total number of months included in the dataset

#use panda to read csv. Import pandas library

import pandas as pd
from pathlib import Path

#filepath
filepath = Path('../CSV/budget_data.csv')

#output text file
output_file_path = Path('../CSV/financial_analysis.txt')

dataframe = pd.read_csv(filepath)
display(dataframe)

# calculate total number of months
def calculate_total_months(filepath):
    total_months = dataframe['Date'].nunique()
    
    return total_months

print("Financial Analysis\n__________________________")

total_months = calculate_total_months(filepath)

if total_months is not None:
    
    print(f"Total number of months in the dataset is : {total_months}")
    

# calculate net total amount of Profit/Losses over the entire period
def calculate_net_total(filepath):
    net_total = dataframe['Profit/Losses'].sum();
    
    return net_total
net_total = calculate_net_total(filepath)

if net_total is not None:
   
    print(f"net total is : ${net_total}")
    
# calculate the average of the changes in PnL over the entire period
def calculate_average_change(filepath):
    dataframe['Change'] = dataframe['Profit/Losses'].diff()
    average_pnl_change = dataframe['Change'].mean()
    return average_pnl_change

average_pnl_change = calculate_average_change(filepath)
if average_pnl_change is not None:
    print(f"average change in Profit/Loss is: {average_pnl_change}")
   # print(f"{dataframe.describe()}")

#calculate the greatest increase in profits (date and amount) over the entire period
def calculate_greatest_increase(filepath):
    
    #get the changes in Profit/Losses
    dataframe['Change'] = dataframe['Profit/Losses'].diff()
    
    #get the row with the greatest increase in profits
    greatest_increase = dataframe.loc[dataframe['Change'].idxmax()]
    
    #get date and amount from greatest increase
    date_increase = greatest_increase['Date']
    amount_increase = greatest_increase['Change']
    
    return date_increase, amount_increase

date_increase, amount_increase = calculate_greatest_increase(filepath)
    
print(f" greatest increase date is : {date_increase} greatest amount is (${amount_increase})")

# calculate the greatest decrease in losses (date and amount) over the entire period

def calculate_greatest_decrease(filepath):
    #get the changes in Profit/Losses
    dataframe['Change'] = dataframe['Profit/Losses'].diff()
    
    #get the row with the greatest decrease in losses
    greatest_decrease = dataframe.loc[dataframe['Change'].idxmin()]
    
     #get date and amount from greatest decrease in losses
    date_decrease = greatest_decrease['Date']
    amount_decrease = greatest_decrease['Change']
    
    return date_decrease, amount_decrease

date_decrease, amount_decrease = calculate_greatest_decrease(filepath)

if date_decrease is not None and amount_decrease is not None:

    print(f" greatest decrease in losses date is : {date_decrease} greatest decrease in loss amount is (${amount_decrease})")

#Export results to a text file
with open(output_file_path, 'w') as file:
            file.write('Financial Analysis\n__________________________\n')
            file.write(f'Total Months: {total_months}\n')
            file.write(f'Net Total Amount of Profit/Losses: ${net_total}\n')
            file.write(f'Average of Changes in Profit/Losses: ${average_pnl_change}\n')
            file.write(f'Greatest Increase in Profits: {date_increase} (${amount_increase})\n')
            file.write(f'Greatest Decrease in Losses: {date_decrease} (${amount_decrease})\n')

            print(f'Results exported to {output_file_path}')
    
    
```
