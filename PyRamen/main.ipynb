```python
    # PyRamen

    #Read in menu_data.csv and set its contents to a separate list object

    #import pandas as pd 

    import csv
    from pathlib import Path

    menu_file_path = Path("../CSV/menu_data.csv")
   

    #initialize an empty list to hold menu_data
    menu_list = []

    #use a with statement to open menu_data
    with open(menu_file_path, 'r') as file:
        menu_reader = csv.reader(file)

        #iterate through rows and append to menu list
        for row in menu_reader:
            menu_list.append(row)

        print(f'menu list: {menu_list}')


    #have to use the with open again because code was throwing an exception about file closed
    with open(menu_file_path, 'r') as file:
        menu_reader = csv.reader(file)

        # use the next function to skip the header (first row of the CSV)
        header = next(menu_reader)

        print(f'\n--------------------\n Header: {header} \n--------------------')

        #Loop over the rest of the rows and append every row to the menu list object
        for row in menu_reader:
            menu_list.append(row)

        print(f'new menu list: {menu_list}')

    #Set up the same process to read in sales_data.csv. However, instead append every row of the sales data to a new sales list object.
    sales_file_path = Path("../CSV/sales_data.csv")

    sales_list = []

    with open(sales_file_path, 'r') as file:
        sales_reader = csv.reader(file)

        #append every row of the sales data to sales list
        for row in sales_reader:
            sales_list.append(row)


    #print(sales_list)
    #can't print the sales_list. Seems the computer runs out of memory
    #IOPub data rate exceeded.
    #The Jupyter server will temporarily stop sending output
    #o the client in order to avoid crashing it.
    #To change this limit, set the config variable
    #`--ServerApp.iopub_data_rate_limit`.

    #Part 2

    #initialize an empty report dictionary
    report_dictionary = {}

    #loop through every row in the sales list object
    for sales_row in sales_list:
            quantity = sales_row[3]
            sales_item= sales_row[4]

            if sales_item not in report_dictionary:
                report_dictionary[sales_item] = {"01-count": 0,"02-revenue": 0,"03-cogs": 0,"04-profit": 0}


            #loop through menu row in menu list

            for menu_row in menu_list:
                item = menu_row[0]
                #category = menu_row[1]
                price = menu_row[3]
                cost = menu_row[4]
                
                #check if item in menu is same as sales_item
                if sales_item == item:
                   
                    profit = int(price) - int(cost)
                    # print(f"profit:{profit} price: {price} cost: {cost}")
                    
                    #add values to the report dictionary created
                    report_dictionary[sales_item]["01-count"] += int(quantity)
                    report_dictionary[sales_item]["02-revenue"] += int(price) * int(quantity)
                    report_dictionary[sales_item]["03-cogs"] += int(cost) * int(quantity)
                    report_dictionary[sales_item]["04-profit"] += int(profit) * int(quantity)
                    
                    break
            else:
                print(f"{sales_item} does not equal {item}! NO MATCH!")
  

    #write out the content of the report dictionary to a text file
    menu_output_file_path = Path("../CSV/output_report.txt")
    with open (menu_output_file_path, 'w') as file:
         for item, metrics in report_dictionary.items():
            file.write(f"{item} {metrics}\n")
        



```
